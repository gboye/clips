```python
import math
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_style("whitegrid")
```

```python
k=10000.0
percentile=.5
bas=[]
haut=[]
result=[]
moyenne=[]
sTete=[]
sQueue=[]
for x in range(1,10000):
    somme=0
    tete=0
    queue=0
    n=100*x
    for i in range(1,n+1):
        somme+=k/i
        if i < percentile*n :
            tete+=k/i
        else:
            queue+=k/i
    b=k*math.log(n+1)
    h=k*(math.log(n+1)+1)
    result.append(somme)
    moyenne.append((b+h)/2)
    bas.append(b)
    haut.append(h)
    sTete.append(tete)
    sQueue.append(queue)
```

```python
plt.plot(sTete)
plt.plot(sQueue)
```

```python
ax=plt.plot(bas[500:],label="bas")
ax=plt.plot(haut[500:],label="haut")
ax=plt.plot(result[500:],label="result")
ax=plt.plot(moyenne[500:],label="moyenne haut-bas")
# plt.yscale('log')
# plt.xscale('log')
plt.legend()

plt.show()
```

```python
a = 2.3
n = 100000000
s = np.random.zipf(a, size=n)

count = np.bincount(s)
k = np.arange(1, s.max() + 1)
print (len(count))
```

```python
fig = plt.figure(figsize=(2500, 5))
plt.bar(k, count[1:], alpha=1.0, label='sample count')
plt.plot(k, n*(k**-a)/zeta(a), 'k.-', alpha=.2,
         label='expected count')   
plt.semilogy()
plt.grid(alpha=0.4)
plt.ylim((.1,n))
plt.xlim((1,len(count)+1))
plt.legend()
plt.title('Zipf sample, a={}, size={}'.format(a,n))
plt.show()
```
